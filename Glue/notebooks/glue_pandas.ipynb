```python
import pandas as pd
from datetime import datetime
```

```python
input_path="sample_data/retail_db/order_items/part-00000-order_items.csv"
```

```python
# Read the input CSV file
df = pd.read_csv(input_path)
df.head(5)
```

```python
# 1. Add a discount column (e.g., 10% discount on subtotal for demonstration purposes)
df['discounted_price'] = (df['order_item_subtotal'] * 0.9).round(2)
df.head()
```

```python
# 2. Add a new column with ingestion timestamp
df['ingestion_timestamp'] = datetime.now().strftime('%Y-%m-%d %H:%M:%S')
df.head(5)
```

```python
# Remove the column 'order_item_product_price'
df = df.drop(columns=['order_item_product_price'])
```

```python
# Write the transformed DataFrame to the output path
# Save the data in a single CSV file, ensuring overwrite mode
output_path ="sample_data/retail_db/order_items/op.csv"
df.to_csv(output_path, index=False)
```
